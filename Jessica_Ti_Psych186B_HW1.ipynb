```python
# Solution for the third problem

  #Part I: Basic Properties and Visualization (7 points)
#(a) (1 point) Given that these vectors are normalized, explain geometrically what their dot product represents.
    # Solution:
    # The dot product of two normalized vectors represents the cosine of the angle between them.
    # A dot product of 1 means they point in the same direction. cos 360 = 1
    # A dot product of 0 means they are perpendicular. cos 90 = 0
    # A dot product of -1 means they point in opposite directions. cos 180 = -1
    # As the vectors are normalized, the dot product depends purely on their relative orientation

#(b) (4 points) Create a histogram of dot products between pairs of such vectors for dimensions: 10, 20, 50, 100, 250, 500, 1000, and 2000.
              # Report mean and standard deviation for each case.

#  Generate normalized (length=1) random vectors with zero mean elements from a uniform distribution.
  # Sum all the elements of your vector and divide by the number of elements, result is zero

    #a. Generating random elements from a uniform distribution.
      # For the vector to be truly random, we need each # within it to have roughly equal chance to get picked hence uniform

import numpy as np
import matplotlib.pyplot as plt

def generate_normalized_vector(n):
    """
    Generate a normalized (length = 1) random vector with zero mean elements from a uniform distribution.

    Parameters:
        n (int): The size of the random vector.

    Returns:
        np.ndarray: The normalized vector.
    """
    # Step 1: Generate random vector from a uniform distribution
    random_vector = np.random.uniform(-1, 1, size=n)

    # Step 2: Compute the mean of the vector
    mean_value = np.mean(random_vector)

    # Step 3: Shift the vector to have zero mean
    zero_mean_vector = random_vector - mean_value

    # Step 4: Calculate the magnitude (norm) of the zero-mean vector
    v_magnitude = np.sqrt(np.sum(zero_mean_vector ** 2))

    # Step 5: Normalize the vector (divide by its magnitude)
    normalized_vector = zero_mean_vector / v_magnitude

    return normalized_vector

def create_histogram_of_dot_products(dimensions):
    """
    Create histograms of dot products between pairs of normalized vectors for specified dimensions.

    Parameters:
        dimensions (list): List of dimensions to generate vectors and calculate dot products.
    """
    for dim in dimensions:
        dot_products = []  # Store dot products for this dimension

        # Generate 100 pairs of random vectors for each dimension - need to run 100 to calculate mean and s.d
        for _ in range(100):
            vec1 = generate_normalized_vector(dim)
            vec2 = generate_normalized_vector(dim)

            # Calculate dot product between the two vectors
            dot_product = np.dot(vec1, vec2)
            dot_products.append(dot_product)

        # Calculate mean and standard deviation of dot products
        mean_dot = np.mean(dot_products)
        std_dot = np.std(dot_products)
        print(f"Dimension {dim}: Mean={mean_dot}, Std Dev={std_dot}")

        # Plot histogram
        plt.hist(dot_products, bins=30, label=f'Dim={dim}')

    # Customize and show the plot
    plt.xlabel("Dot Product")
    plt.ylabel("Frequency")
    plt.title("Histogram of Dot Products for Various Dimensions")
    plt.legend()
    plt.show()

# Define dimensions for the experiment
dimensions = [10, 20, 50, 100, 250, 500, 1000, 2000]

# Create histograms of dot products
create_histogram_of_dot_products(dimensions)

#(c) (2 points) The mean of these dot product distributions has a theoretical value. Determine this value and explain why it makes sense geometrically.
              # Compare with your simulation results.
```

```python
# Track A Step 2

import numpy as np

# Step 2: Simulate normalized random vectors and compute dot products
n_samples = 100000  # Number of random vector pairs
dot_products = [] # Create a list so we can convert it to a numpy arraay later to make calculating mean and variance easier

for _ in range(n_samples):
    # Generate two random 2D vectors
    v1 = np.random.uniform(-1, 1, 2)
    v2 = np.random.uniform(-1, 1, 2)

    # Why valid: Drawn from uniform distribution between -1 and 1. Component of vectors randomly chosen from this range, mimics randomness in real-world vector selection.

    # Normalize the vectors
    v1_normalized = v1 / np.linalg.norm(v1)
    v2_normalized = v2 / np.linalg.norm(v2)

    # Compute and store the dot product
    dot_products.append(np.dot(v1_normalized, v2_normalized))

    # Why valid: Storing dot product of each pair of vectors to later analyze the distribution of these dot products.
    # Using list as it is an ordered collection, keeping track of each individual result from the simulation.


# Convert to numpy array for calculations
dot_products = np.array(dot_products)

# Calculate mean, squared mean, and variance
E_X = np.mean(dot_products)
# Why valid: E(X) represents mena, average similarity between the two vectors over all the random pairs. Gives sense of  typical alignment between two random normalized vectors.
E_X_squared = np.mean(dot_products**2)
# Why valid: By squaring the values,  focusing more on the spread of the values, irrespective of their direction.
variance = E_X_squared - E_X**2

print("E[X]:", E_X)
print("E[X^2]:", E_X_squared)
print("Variance:", variance)
```

```python
# Step 3: Simulate dot products in higher dimensions
dims = [2, 3, 5, 10, 20, 50, 100]  # List of dimensions incl 3D to experiment with effect
results = {}

for d in dims:
    dot_products = []
    for _ in range(n_samples):
        # Generate two random vectors in d dimensions
        v1 = np.random.uniform(-1, 1, d)
        v2 = np.random.uniform(-1, 1, d)

        # Normalize the vectors
        v1_normalized = v1 / np.linalg.norm(v1)
        v2_normalized = v2 / np.linalg.norm(v2)

        # Compute and store the dot product
        dot_products.append(np.dot(v1_normalized, v2_normalized))

    # Convert to numpy array for calculations
    dot_products = np.array(dot_products)

    # Calculate mean, squared mean, and variance
    E_X = np.mean(dot_products)
    E_X_squared = np.mean(dot_products**2)
    variance = E_X_squared - E_X**2

    # Store results for this dimension
    results[d] = {
        "Mean": E_X,
        "E[X^2]": E_X_squared,
        "Variance": variance
    }

# Display results for all dimensions
for d, stats in results.items():
    print(f"Dimension {d}: Mean = {stats['Mean']:.4f}, Variance = {stats['Variance']:.4f}") # Corrected to 4 d.p so not overly long

# Pattern for higher dimensions:
# As the dimension increases, the variance decreases.
# This makes sense because as the number of dimensions increases, the possible directions that a vector can point to becomes more abundant.
# So in lower dimensions like 2D or 3D, vectors tend to align or oppose each other in more noticeable ways. As a result, the dot product can vary more widely.
# But for higher dimensions, vectors are spread out more uniformly. Since the vectors are normalized, they are more likely to be close to orthogonal to each other, so its closer to zero.
```

```python
# Solution for the fourth problem

# (a) Assume a square with side 2 units long centered on zero, that is, the sides run from -1 to +1 along both axes. The area of this square is 4.
# (b) Compute two random number (x,y) uniformly distributed between -1 and +1. These two numbers (x,y) correspond to a point in the square.

import numpy as np

# (a) The area of the square is 4, we are generating points uniformly within this square.
n_samples = 100000  # Number of points to simulate
points_in_circle = 0  # To keep track of how many points fall inside the circle

for _ in range(n_samples):
    # (b) Generate two random numbers (x, y) between -1 and 1
    x, y = np.random.uniform(-1, 1, 2)

    # (d) Check if the point is inside the circle
    if x**2 + y**2 <= 1:  # If inside the unit circle
        points_in_circle += 1

# (e) Do this many many times. Keep track of how many points fall within the circle.
# (f) If the (x,y) are chosen from a truly uniform distribution the ratio of total points to the points lying in the circle is the ratio of their areas, that is, 4/pi. This ratio therefore gives an estimate of pi.
#(g) Try it out and describe the results.

# (e) Calculate the ratio of points inside the circle
ratio_inside_circle = points_in_circle / n_samples

# Estimate Pi
estimated_pi = 4 * ratio_inside_circle

print(f"Estimated Pi: {estimated_pi}")
print(f"Points inside circle: {points_in_circle}")
print(f"Total points: {n_samples}")

# (f) The estimated Pi I got was 3.1488, where 78720/100000 of the points fell within the circle.

```
